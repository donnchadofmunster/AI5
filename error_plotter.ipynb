```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
file_path = 'error_data.csv'
df = pd.read_csv(file_path, header=None, names=['Timestamp', 'Value'])

# Ensure 'Value' is numeric
df['Value'] = pd.to_numeric(df['Value'], errors='coerce')

# Drop NaN values and filter out values at 160
df = df.dropna()
df_filtered = df[df['Value'] < 160]

# Calculate average ignoring values at 160
average_value = df_filtered['Value'].mean()
print(f'Average Value (excluding 160): {average_value}')

# Limit to first 2500 entries
df_limited = df_filtered.iloc[:2500]

# Plot with seaborn
sns.set_theme(style="darkgrid")
plt.figure(figsize=(10, 5))
sns.lineplot(data=df_limited, x=df_limited.index, y='Value', linewidth=2, markersize=8)
plt.xlabel('Index')
plt.ylabel('Absolute of Error')
plt.title('Error against Index')
plt.grid()
plt.show()
```
